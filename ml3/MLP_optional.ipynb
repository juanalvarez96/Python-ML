```python
# Load datasets
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes = True)

from scipy.special import fresnel
```

```python
n_examples = 500
x_min = 0
x_max = 3
fresnel_func = 0
```

```python
# Generate dataset
x = np.linspace(x_min, x_max, num=n_examples)
y = fresnel(x)[fresnel_func]

X=x.reshape(-1,1)
plt.plot(x,y)
plt.show()
```

```python
# Make MLP learn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.neural_network import MLPRegressor

# Split dataset
X_train, X_test, y_train, y_test = \
train_test_split(X,y,test_size = 0.3, random_state=44)
```

```python
# Train the network (modify)
mlp = MLPRegressor(hidden_layer_sizes=(1000,753,706, 402, 293, 100, 60, 15))
mlp.fit(X_train, y_train)
```

```python
# Predict
preds = mlp.predict(X_test)
plt.plot(x,y)
plt.scatter(X_test, preds, color = 'g')
plt.show()

print('Score:', r2_score(y_test, preds))
```
